```python
#import the requried libraries
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import math
```

```python
#import the csv file of endorsements
df_raw = pd.read_csv('endorsements-2020.csv')

#drop rows that contain null values for endorsees
df = df_raw[df_raw['endorsee'].notnull()]

#drop the source column to clean up the look of the dataframe
df.drop(columns=['source'],inplace=True)

#show the resulting dataframe
df
```

```python
df.sort_values(['endorsee'],ascending=True,inplace=True)
#create a list of unique endorsees
z = list(df['endorsee'].unique())
print(z)
```

```python
#find and print the total number of available endorsement points
print("Total number of available points:",df_raw['points'].sum(axis=0))

#find and print the number of endorsement points each candidate has recieved
for x in range(len(z)):
    df_i = df.loc[df['endorsee'] == (z[x])]
    print("Point total of",z[x],":",df_i['points'].sum(axis = 0))
```

```python
df.sort_values(['points'],ascending=False,inplace=True)
df
```
